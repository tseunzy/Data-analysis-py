```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'D:\Git_repo\Data-analysis-py\Numpy\students_messy_120k.csv')
```

```python
df.dtypes
df.shape
```

```python
# YearsCode to numerical
df['YearsCode'].unique()
```

```python
df['YearsCode'].replace({'Less than 1 year': '0', 'More than 50 years': '53', 'five': '5'}, inplace=True)
```

```python
df['YearsCode'] = pd.to_numeric(df['YearsCode'], errors='coerce')
```

```python
df.dtypes
```

```python
df['ClassLevel'].unique()
```

```python
# ClassLevel
df['ClassLevel'] = df['ClassLevel'].replace({'SS-3': "SS3"}).str.strip().str.upper().astype("string")
```

```python
df['ClassLevel'].dtypes
```

```python
# country 
df['Country'] = df['Country'].astype('string').str.strip().str.title()
```

```python
df['Gender'].unique()
```

```python
#  gender, convert all to upper and replace 
df['Gender'] = df['Gender'].astype('string').str.strip().str.upper().replace({'MALE': 'M', 'FEMALE': 'F', '': pd.NA})
```

```python
df
```

```python
df['Date'].unique()
```

```python
# Date 
 
d1 = pd.to_datetime(df['Date'], errors='coerce')
d3 = pd.to_datetime(df['Date'], format='%d/%m/%Y %I:%M %p', errors='coerce')
df['Date'] = d1.fillna(d3)
```

```python
df['Date']
```

```python
df['MathScore'].unique()
```

```python
# clip MathScore, EnglishScore scores to 0–100 and fill missing with column mean
# remove negative scores and scores higher than 100
for mateng in ['MathScore', 'EnglishScore']:
    df[mateng] = pd.to_numeric(df[mateng], errors='coerce')
    df[mateng] = df[mateng].clip(0, 100)
    df[mateng] = df[mateng].fillna(df[mateng].mean())
```

```python
# Fix Age
# Take out negative ages and ages thats seems too high manually
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df.loc[(df['Age'] < 10) | (df['Age'] > 22), 'Age'] = np.nan
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
# AttendancePct clip it to 0, 100
# sort out attendent for missing values
df['AttendancePct'] = (pd.to_numeric(df['AttendancePct'], errors='coerce')).clip(0, 100)
df['AttendancePct'] = df['AttendancePct'].fillna(df['AttendancePct'].mean())
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df.to_csv('students_cleaned.csv', index=False, encoding='utf-8-sig')
```

```python
# Weekly resample summary
# # Keep only rows with valid dates
dp_date = df.dropna(subset=['Date'])
dp_date = dp_date.set_index('Date').sort_index()
```

```python
dp_date.head()
```

```python
# Weekly mean
weekly = dp_date.resample('W').agg({'MathScore': 'mean', 'EnglishScore': 'mean', 'AttendancePct': 'mean', 'Age': 'mean', 'YearsCode': 'mean', 'StudentID': 'count'})
```

```python
weekly = weekly.rename(columns={'StudentID': 'RowsInWeek'})
```

```python
weekly
```

```python
weekly.to_csv('CLEAN_WEEKLY.csv', index=True, encoding="utf-8-sig")
```
